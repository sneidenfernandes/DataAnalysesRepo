### Programming Language Popularity

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('sof_results.csv')
```

#### Language Users Plot :

```python
languages = df['LanguageHaveWorkedWith'].str.split(';', expand=True)
languages = languages.stack().value_counts()
languages.plot(kind='bar', color='green', figsize=(15,8))
```
